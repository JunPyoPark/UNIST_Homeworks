# csv to networkx & Gephi

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Ulsan_apart.csv', encoding='cp949')
```

```python
data
```

```python
# 평당 가격 평균
data['PR_per_PY'] = data['PR_SUM'] / data['PY_SUM']
```

```python
data.iloc[0]
```

```python
price_apart = data.sum().iloc[14:-3]
```

```python
plt.figure(1, figsize=(14, 7))# control figure size
plt.bar(price_apart.index, price_apart)
plt.title('Apart Price Distribution in Ulsan', fontsize=14);
plt.show()
```

```python
data = data.sort_values(by=['latitude','longitude'])
```

```python
data.columns
```

```python
# data.where(data['latitude'] < 35.43).dropna().where(data['longitude'] < 129.29).dropna()
```

```python
G = nx.Graph()

for idx,row in data.iterrows():
    G.add_node(row['Unnamed: 0'],Label=idx,latitude=row['latitude'], longitude=row['longitude'],
              pr_05u=row['PR_05U'], pr_05=row['PR_05'], pr_1=row['PR_1'], pr_2=row['PR_2'], pr_3=row['PR_3'],pr_4=row['PR_4'],
              pr_5=row['PR_5'],pr_6=row['PR_6'],pr_7_Over=row['PR_7_O'],avg_price = row['PR_per_PY'])
```

```python
print(nx.info(G))
```

```python
nx.write_gexf(G,'apart.gexf')
```

```python
import pandas as pd
import folium
```

```python
starbucks = pd.read_csv('Ulsan_STARBUCKS_list.csv')
```

```python
starbucks.head()
```

```python
def draw_map(name,iter_num):
    data = pd.read_csv(name + '.csv')
    data = data.drop(['Unnamed: 0'],axis=1)
    data = data.sort_values(by='prob', ascending=False)
    map_osm = folium.Map(location=(35.550548,129.298364), zoom_start=13)

    for ix, row in starbucks.iterrows():
        location = (row['lat'], row['lot'])
        folium.Circle(
            location=location,
            radius=500,
            color='green',
            weight=1,
            fill_opacity=0.6,
            opacity=1,
            fill_color='green',
            fill=True,  # gets overridden by fill_color
            popup=str(row['s_name'] + '점')
        ).add_to(map_osm)
    count=0
    for ix, row in data.iterrows():
        location = (row['latitude'], row['longitude'])
        folium.Circle(
            location=location,
            radius=500 * row['prob'],
            color='red',
            weight=1,
            fill_opacity=0.6,
            opacity=1,
            fill_color='red',
            fill=True,  # gets overridden by fill_color
            # popup=str(row['prob'])
        ).add_to(map_osm)
        count += 1
        if(count==iter_num):
            break

    unist_loc = (35.573830,129.190944)
    folium.Marker(unist_loc, popup='UNIST').add_to(map_osm)
    return map_osm
```

# Whole Plot with NO_PCA

```python
draw_map('NO_PCA_result',-1)
```
